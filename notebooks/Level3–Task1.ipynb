# Level 3 - Task 1: Restaurant Reviews Analysis

## Objective

The objective of this task is to analyze restaurant ratings and customer reviews.

### This task includes:

- Analyze rating text distribution.
- Calculate average ratings for each rating category.
- Visualize rating text frequencies.
- Draw business insights.

### Libraries Used

- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv("../Dataset.csv")
```

## Step 1: Check Rating Text Distribution

The frequency of each rating category is calculated.

```python
rating_text = df["Rating text"].value_counts()

rating_text
```

## Step 2: Create Summary Table

Display the count of restaurants in each rating category.

```python
summary = pd.DataFrame({
    "Rating Text": rating_text.index,
    "Restaurant Count": rating_text.values
})

summary
```

## Step 3: Visualize Rating Text Distribution

A bar chart is used to visualize customer review categories.

```python
os.makedirs("../images", exist_ok=True)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    data=summary,
    x="Restaurant Count",
    y="Rating Text"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Restaurant Rating Categories")
plt.xlabel("Restaurant Count")
plt.ylabel("Rating Category")

plt.savefig("../images/rating_text_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

## Step 4: Average Rating for Each Rating Category

Calculate the average aggregate rating for each rating text.

```python
avg_rating = (
    df.groupby("Rating text")["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
)

avg_rating
```

## Step 5: Visualize Average Ratings

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=avg_rating.values,
    y=avg_rating.index
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.title("Average Rating by Rating Category")
plt.xlabel("Average Rating")
plt.ylabel("Rating Category")

plt.savefig("../images/rating_category_average.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

# Business Insight

- Restaurants with higher ratings attract more customers.
- "Excellent" restaurants represent the highest-performing businesses.
- Restaurants with lower ratings should focus on improving service quality.
