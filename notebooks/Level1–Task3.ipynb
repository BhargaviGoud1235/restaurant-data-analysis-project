# Level 1 - Task 3: Price Range Distribution

## Objective

The objective of this task is to analyze the distribution of restaurant price ranges.

### This task includes:

- Visualizing the distribution of restaurant price ranges.
- Calculating the percentage of restaurants in each price range.
- Drawing conclusions from the distribution.

### Libraries Used

- Pandas
- Matplotlib
- Seaborn

## Step 1: Count Restaurants in Each Price Range

The number of restaurants in each price range category is calculated.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

```python
df = pd.read_csv("../Dataset.csv")

# Display first five rows
df.head()
```

```python
price_counts = df["Price range"].value_counts().sort_index()

price_counts
```

## Step 2: Calculate Percentage Distribution

The percentage of restaurants in each price range is calculated.

```python
price_percentage = (price_counts / len(df)) * 100

price_percentage.round(2)
```

## Step 3: Create Summary Table

A summary table is created showing the number and percentage of restaurants in each price range.

```python
summary = pd.DataFrame({
    "Price Range": price_counts.index,
    "Restaurant Count": price_counts.values,
    "Percentage (%)": price_percentage.values.round(2)
})

summary
```

## Step 4: Visualize the Distribution

A bar chart is created to visualize the restaurant distribution across different price ranges.

```python
import os

# Create images folder if it doesn't exist
os.makedirs("../images", exist_ok=True)

plt.figure(figsize=(8,5))

sns.barplot(
    data=summary,
    x="Price Range",
    y="Restaurant Count"
)

plt.title("Distribution of Restaurant Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.savefig("../images/price_range_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

## Step 5: Pie Chart Representation

A pie chart is created to show the percentage distribution of restaurants across price ranges.

```python
plt.figure(figsize=(7,7))

plt.pie(
    summary["Restaurant Count"],
    labels=summary["Price Range"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Restaurant Price Range Distribution")

plt.savefig("../images/price_range_piechart.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

# Conclusion

## Key Findings

- The restaurant dataset contains four price range categories.
- The majority of restaurants fall into the lower and medium price ranges.
- Higher price range restaurants are comparatively fewer.
- The bar chart and pie chart clearly illustrate the distribution of restaurant price categories.

This analysis helps understand customer affordability and restaurant pricing trends.
