# Level 2 - Task 3: Geographic Analysis

## Objective

The objective of this task is to analyze the geographical distribution of restaurants using their latitude and longitude coordinates.

### This task includes:

- Visualizing restaurant locations using a scatter plot.
- Identifying cities with the highest restaurant concentration.
- Understanding the geographical spread of restaurants.

### Libraries Used

- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load Dataset
df = pd.read_csv("../Dataset.csv")
```

## Step 1: Check Geographic Coordinates

The latitude and longitude columns are examined to ensure that restaurant locations are available for analysis.

```python
df[["Latitude", "Longitude"]].head()
```

## Step 2: Check Missing Geographic Data

Verify whether any restaurant records are missing latitude or longitude values.

```python
df[["Latitude", "Longitude"]].isnull().sum()
```

## Step 3: Visualize Restaurant Locations

A scatter plot is used to display restaurant locations based on their latitude and longitude coordinates.

```python
os.makedirs("../images", exist_ok=True)

plt.figure(figsize=(10,8))

plt.scatter(
    df["Longitude"],
    df["Latitude"],
    alpha=0.5,
    s=15
)

plt.title("Geographical Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.savefig(
    "../images/geographical_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Step 4: Identify Cities with the Highest Restaurant Count

The restaurants are grouped by city to determine where restaurant density is highest.

```python
city_distribution = df["City"].value_counts()

city_distribution.head(10)
```

## Step 5: Create Summary Table

The following table displays the top 10 cities based on restaurant count.

```python
summary = pd.DataFrame({
    "City": city_distribution.head(10).index,
    "Restaurant Count": city_distribution.head(10).values
})

summary
```

## Step 6: Visualize Restaurant Density by City

The following bar chart compares the top 10 cities with the highest number of restaurants.

```python
plt.figure(figsize=(10,6))

ax = sns.barplot(
    data=summary,
    x="Restaurant Count",
    y="City"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Top 10 Cities by Restaurant Count")
plt.xlabel("Restaurant Count")
plt.ylabel("City")

plt.savefig(
    "../images/top10_cities_restaurants.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Step 7: Identify the City with the Highest Restaurant Density

Determine which city contains the largest number of restaurants.

```python
top_city = city_distribution.idxmax()
top_count = city_distribution.max()

print("City with Highest Restaurant Density:", top_city)
print("Number of Restaurants:", top_count)
```
