# Level 3 - Task 3: Price Range vs Online Delivery & Table Booking Analysis

## Objective

The objective of this task is to examine the relationship between restaurant price range and service availability.

### This task includes:

- Analyze online delivery across different price ranges.
- Analyze table booking across different price ranges.
- Visualize the distribution.
- Draw business insights.

### Libraries Used

- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load Dataset
df = pd.read_csv("../Dataset.csv")
```

## Step 1: Explore Price Range

Display the distribution of restaurants across different price ranges.

```python
price_counts = df["Price range"].value_counts().sort_index()

price_counts
```

## Step 2: Online Delivery by Price Range

Analyze the availability of online delivery across different price ranges.

```python
online_delivery = pd.crosstab(
    df["Price range"],
    df["Has Online delivery"]
)

online_delivery
```

## Step 3: Visualize Online Delivery by Price Range

```python
os.makedirs("../images", exist_ok=True)

online_delivery.plot(
    kind="bar",
    figsize=(8,6)
)

plt.title("Online Delivery by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.savefig(
    "../images/online_delivery_price_range.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Step 4: Table Booking by Price Range

Analyze the availability of table booking across different price ranges.

```python
table_booking = pd.crosstab(
    df["Price range"],
    df["Has Table booking"]
)

table_booking
```

## Step 5: Visualize Table Booking by Price Range

```python
table_booking.plot(
    kind="bar",
    figsize=(8,6)
)

plt.title("Table Booking by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.savefig(
    "../images/table_booking_price_range.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Step 6: Average Rating by Price Range

```python
avg_rating = (
    df.groupby("Price range")["Aggregate rating"]
    .mean()
    .round(2)
)

avg_rating
```

## Step 7: Visualize Average Ratings

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.savefig(
    "../images/average_rating_price_range.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

# Business Insight

- Higher-priced restaurants are more likely to offer table booking services.
- Online delivery is commonly available in lower and medium-priced restaurants.
- Average ratings vary across different price ranges, providing insight into customer satisfaction.
- These findings help restaurants decide which services to offer based on their pricing strategy.
